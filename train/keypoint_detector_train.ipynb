```python
def masked_l2_loss(preds, target_xy, vis):
    mask = vis.unsqueeze(-1)                              # (B,14,1)
    diff = (preds - target_xy) * mask                     
    sq = diff.pow(2)                                      # (B,14,2)
    per_kpt = sq.sum(dim=-1)                              # (B,14)
    denom = mask.squeeze(-1).sum(dim=1).clamp(min=1)      # (B,)
    per_sample = per_kpt.sum(dim=1) / denom               # (B,)
    return per_sample.mean()
```

```python
import torch
import torch.nn as nn
from torchvision.models import resnet50, ResNet50_Weights

class SEBlock(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.fc = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(channels, channels // reduction, 1, bias=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(channels // reduction, channels, 1, bias=True),
            nn.Sigmoid()
        )

    def forward(self, x):
        scale = self.fc(x)
        return x * scale

class KeypointFPNRegressor(nn.Module):
    def __init__(self, num_keypoints=14):   #   14
        super().__init__()
        self.num_keypoints = num_keypoints

        self.backbone = resnet50(weights="ResNet50_Weights.IMAGENET1K_V1")

        # feature
        self.layer1 = nn.Sequential(*list(self.backbone.children())[:5])  # (B,256,H/4,W/4)
        self.layer2 = self.backbone.layer2                                # (B,512,H/8,W/8)
        self.layer3 = self.backbone.layer3                                # (B,1024,H/16,W/16)
        self.layer4 = self.backbone.layer4                                # (B,2048,H/32,W/32)

        # 
        self.up2 = nn.ConvTranspose2d(2048, 512, kernel_size=2, stride=2)             # H/32H/16
        self.up1 = nn.ConvTranspose2d(512 + 1024, 256, kernel_size=2, stride=2)       # H/16H/8

        self.fusion_conv = nn.Sequential(
            nn.Conv2d(256 + 512, 256, kernel_size=3, padding=1, bias=False),
            nn.ReLU(inplace=True),
            SEBlock(256),
            nn.AdaptiveAvgPool2d((1, 1)),
        )

        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_keypoints * 2),
            nn.Sigmoid()  
        )

    def forward(self, x):
        x1 = self.layer1(x)
        x2 = self.layer2(x1)
        x3 = self.layer3(x2)
        x4 = self.layer4(x3)

        x = self.up2(x4)                 # (B,512,H/16,W/16)
        x = torch.cat([x, x3], dim=1)    # (B,512+1024,H/16,W/16)
        x = self.up1(x)                  # (B,256,H/8,W/8)
        x = torch.cat([x, x2], dim=1)    # (B,256+512,H/8,W/8)

        x = self.fusion_conv(x)          # (B,256,1,1)
        out = self.head(x)               # (B, 14*2)
        return out.view(-1, self.num_keypoints, 2)  # (B,14,2)
```

```python
from tqdm import tqdm
import matplotlib.pyplot as plt
import os

import os
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt

def train(model, optimizer, train_loader, val_loader, device, num_epochs=100, patience=5):
    model_save_dir = r"D:\bishe\w3_net_model"
    os.makedirs(model_save_dir, exist_ok=True)

    best_val_loss = float('inf')
    early_stop_counter = 0
    train_losses, val_losses = [], []

    for epoch in range(1, num_epochs + 1):
        # ===== Train =====
        model.train()
        total_train_loss = 0.0

        pbar = tqdm(train_loader, desc=f"Epoch {epoch} | Train", leave=False)
        for imgs, kps, vis in pbar:
            imgs = imgs.to(device, non_blocking=True)      # (B,3,H,W)
            kps  = kps.to(device, non_blocking=True)       # (B,14,2)
            vis  = vis.to(device, non_blocking=True)       # (B,14)

            preds = model(imgs)                             # (B,14,2)
            loss  = masked_l2_loss(preds, kps, vis)        

            optimizer.zero_grad(set_to_none=True)
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item()
            pbar.set_postfix(loss=f"{loss.item():.4f}")

        avg_train_loss = total_train_loss / max(1, len(train_loader))

        # ===== Validate =====
        model.eval()
        total_val_loss = 0.0
        with torch.no_grad():
            pbar_val = tqdm(val_loader, desc=f"Epoch {epoch} | Val  ", leave=False)
            for imgs, kps, vis in pbar_val:
                imgs = imgs.to(device, non_blocking=True)
                kps  = kps.to(device, non_blocking=True)
                vis  = vis.to(device, non_blocking=True)

                preds = model(imgs)
                loss  = masked_l2_loss(preds, kps, vis)
                total_val_loss += loss.item()
                pbar_val.set_postfix(val_loss=f"{loss.item():.4f}")

        avg_val_loss = total_val_loss / max(1, len(val_loader))

        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)
        print(f"[Epoch {epoch:03d}] Train: {avg_train_loss:.4f} | Val: {avg_val_loss:.4f}")

        # ===== Save / Early stop =====
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            early_stop_counter = 0
            torch.save(model.state_dict(), os.path.join(model_save_dir, "best_model.pth"))
            print(" Best model saved.")
        else:
            early_stop_counter += 1
            if early_stop_counter >= patience:
                print(" Early stopping triggered.")
                break

    # ===== Plot & Save final =====
    torch.save(model.state_dict(), os.path.join(model_save_dir, "final_model.pth"))
    plt.figure()
    plt.plot(train_losses, label="Train")
    plt.plot(val_losses, label="Val")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss Curve")
    plt.legend(); plt.grid(True)
    plt.savefig(os.path.join(model_save_dir, "loss_curve.png"))
    plt.show()
```

```python
import os, random
import numpy as np
from PIL import Image, UnidentifiedImageError
from torch.utils.data import Dataset, DataLoader
import torch
import torchvision.transforms as T
from tqdm import tqdm

# ==== Dataset ====
class Avian14Dataset(Dataset):
    def __init__(self, root, transform=None, subset_ratio=1.0):
        self.root = root
        self.sids = sorted([d for d in os.listdir(root) if os.path.isdir(os.path.join(root, d))])
        if subset_ratio < 1.0:
            random.seed(42)
            n = int(len(self.sids) * subset_ratio)
            self.sids = random.sample(self.sids, n)
        
        self.transform = transform

    def __len__(self):
        return len(self.sids)

    def __getitem__(self, idx):
        sid = self.sids[idx]
        d = os.path.join(self.root, sid)
        img_path = os.path.join(d, "image.png")
        kps_path = os.path.join(d, "keypoints14_norm.npy")
        try:
            img = Image.open(img_path).convert("RGB")
        except UnidentifiedImageError:
            raise RuntimeError(f": {img_path}")

        kps = np.load(kps_path).astype(np.float32)  # (14,3) norm
        if self.transform:
            img = self.transform(img)
        return {"image": img, "keypoints": kps, "sid": sid}

# ==== transform ====
transform = T.Compose([
    T.Resize((256, 256)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225])
])

# ==== Evaluation ====
def evaluate_folder(model, root_dir, device, num_kpts=14, subset_ratio=1.0):
    ds = Avian14Dataset(root_dir, transform=transform, subset_ratio=subset_ratio)
    dl = DataLoader(ds, batch_size=32, shuffle=False)

    nvis, n_pck05, n_pck10 = 0, 0, 0
    sum_err = np.zeros(num_kpts, dtype=np.float64)
    cnt_err = np.zeros(num_kpts, dtype=np.int64)

    with torch.no_grad():
        for batch in tqdm(dl, desc="Evaluating"):
            imgs = batch["image"].to(device)
            gt = batch["keypoints"].numpy()   # (B,14,3)

            preds = model(imgs).cpu().numpy() # (B,14,2) in [0,1]
            B,_,_ = gt.shape

            for b in range(B):
                gt_xy = gt[b,:,:2]; vis = gt[b,:,2]
                pred_xy = preds[b]

                gt_px = gt_xy * 256
                pred_px = pred_xy * 256

                dists = np.linalg.norm(pred_px - gt_px, axis=1)  # (14,)
                for j in range(num_kpts):
                    if vis[j] > 0.5:
                        sum_err[j] += dists[j]
                        cnt_err[j] += 1
                        nvis += 1
                        if dists[j] <= 0.05 * 256: n_pck05 += 1
                        if dists[j] <= 0.10 * 256: n_pck10 += 1

    mpjpe = sum(sum_err) / max(nvis,1)
    per_joint = sum_err / np.maximum(cnt_err,1)

    return {
        "samples": len(ds),
        "visible_kpts": nvis,
        "MPJPE": mpjpe,
        "PCK@0.05": n_pck05 / max(nvis,1),
        "PCK@0.10": n_pck10 / max(nvis,1),
        "per_joint_mpjpe": per_joint
    }

# ==== run ====
ROOT_DIR = r"D:\bishe\data_nor_14_w3"
metrics = evaluate_folder(model, ROOT_DIR, device, subset_ratio=0.1)

print("\n=== Evaluation on 10% subset of data_nor_14_w3 ===")
print(f"Samples: {metrics['samples']}")
print(f"Visible keypoints: {metrics['visible_kpts']}")
print(f"MPJPE (px): {metrics['MPJPE']:.3f}")
print(f"PCK@0.05  : {metrics['PCK@0.05']*100:.2f}%")
print(f"PCK@0.10  : {metrics['PCK@0.10']*100:.2f}%")
print("Per-joint MPJPE (px):")
for j, e in enumerate(metrics["per_joint_mpjpe"]):
    print(f"  Joint {j:02d}: {e:.3f}")
```
